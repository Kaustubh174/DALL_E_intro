## Generating original image using DALL-E 3

```python
import openai
import os
os.environ['OPENAI_API_KEY'] = 'your api key'  #put your api key here
openai.api_key = os.getenv('OPENAI_API_KEY')
```

```python
from openai import OpenAI
client = OpenAI()

prompt = 'An Image of Evil AI robots taking over the world and distroying all human civilization'

response = client.images.generate(
    model = 'dall-e-3',
    prompt = prompt,
    style = 'vivid',
    size = '1024x1024',
    quality= 'standard',
    n=1
)
print(response)
```

#### Now we need to get the image through the link

```python
image_url = response.data[0].url
print(image_url)
```

#### we need to convert the url image and display it

```python
import requests
import shutil

image_resource = requests.get(image_url, stream= True)

if image_resource.status_code == 200:
    image_name = "dalle-3image.png"
    with open (image_name,'wb')as f:
        shutil.copyfileobj(image_resource.raw,f)
else:
    print("error")
```
